# Example of field-free orientation of OCS molecule using THz field

This example demonstrates how strong THz pulses (up to 1 MV/cm) can be used to achieve field-free alignment and orientation of the OCS molecule.  

In contrast to the [ocs_thz_orientation.ipynb](ocs_thz_orientation.ipynb) example, the calculations here are carried out for a fixed value of the angular-momentum projection quantum number $m$, which remains a good quantum number when all external fields are applied along the $Z$ axis.

We take the THz pulse profile from Fig. 4 in X. Wu, A.-L. Calendron, K. Ravi, et. al, "Optical generation of single-cycle 10 MW peak power 100 GHz waves", *Optics Express* **24**, 21059-21069 (2016), https://doi.org/10.1364/OE.24.021059.

```python
import csv

import matplotlib.pyplot as plt
import numpy as np
from scipy import constants
from scipy.interpolate import interp1d

from richmol import DIP_X_FIELD_INVCM, CartTensor, RotStates
from richmol.tdse import propagate_expm, propagate_rk, propagate_expokit
```

Define molecule and its properties, compute rotational states and matrix elements.

```python
xyz = (
    "C",  0.0,  0.0,  -0.522939783141,
    "O",  0.0,  0.0,  -1.680839357,
    "S",  0.0,  0.0,  1.037160128,
)

# rotational states
jmax = 10

# Specify the m quantum number via the `m_list` parameter to RotStates.
# Since the field will be applied along the Z-axis, m remains a good quantum number.
# If omitted, the calculation becomes more expensive, including all values m=-j..j for j=0..jmax.

states = RotStates.from_geometry(jmax, xyz, m_list=[0], print_enr=True)

# molecular-frame dipole moment (in au)
dip = [0, 0, -0.31093]

# molecular-frame polarizability tensor (in au)
pol = [
    [25.5778097, 0, 0],
    [0, 25.5778097, 0],
    [0, 0, 52.4651140],
]

# rotational matrix elements of laboratory-frame
# dipole and polarizability tensors
dip = CartTensor(states, dip)
pol = CartTensor(states, pol)
cos_beta = CartTensor(states, "cos(beta)")
cos2_beta = CartTensor(states, "cos^2(beta)")
```

Initialize THz field, use pulse profile provided in file [thz_field.csv](thz_field.csv).

```python
with open("thz_field.csv") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=";")
    _field = np.array(
        [[float(elem.replace(",", ".")) for elem in row] for row in csv_reader]
    )
    _field -= [np.min(_field.T[0]), 0]
    _field /= [1, np.max(_field.T[1])]
    _f_field = interp1d(*_field.T, fill_value=0, bounds_error=False)


def field_z(time, peak_field):
    return _f_field(time) * peak_field


plt.figure(figsize=(6, 4), dpi=300)

t = np.linspace(0, 15, 1000)
plt.plot(t, field_z(t, peak_field=1), linewidth=2)
plt.xlabel("Time (ps)")
plt.ylabel("THz field profile")
plt.grid(True, linestyle=":", alpha=0.6)
plt.yticks(fontsize=10)
plt.tight_layout()
```

Solve TDSE for OCS rotations in THz field.

```python
# THz field along the Z direction
peak_field = 1e6  # 1MV/cm
field_au = constants.value("atomic unit of electric field")
field_func = lambda t: np.array(
    [0, 0, field_z(t, peak_field=peak_field * 100 / field_au)]
)

# initial wavepacket: J=abs(m_list[0]) state (see `m_list` parameter in `RotStates`)
c0 = np.zeros(states.mat().shape[0], dtype=np.complex128)
c0[0] = 1.0

time, c_t = propagate_expokit(
    states,
    [dip],
    [-DIP_X_FIELD_INVCM],
    field_func=field_func,
    t0=0,
    t1=500,
    dt=0.1,
    time_unit="ps",
    c0=c0,
)

# time, c_t = propagate_expm(
#     states,
#     [dip],
#     [-DIP_X_FIELD_INVCM],
#     field_func=field_func,
#     t0=0,
#     t1=500,
#     dt=0.1,
#     time_unit="ps",
#     c0=c0,
# )

# time, c_t = propagate_rk(
#     states,
#     [dip],
#     [-DIP_X_FIELD_INVCM],
#     field_func=field_func,
#     t0=0,
#     t1=500,
#     time_unit="ps",
#     c0=c0,
# )
```

Plot state populations.

```python
# state populations
pop_t = np.abs(c_t) ** 2

# indices of states with nonzero populations
state_ind = np.any(np.abs(pop_t) > 1e-3, axis=0).nonzero()[0]

# select quanta corresponding to the chosen "m" quantum number
state_qua = [
    f"$|J,m,k,\\tau\\rangle$ = $|{j}, {m}, {k}, {tau}\\rangle$"
    for (j, m, k, tau, *_) in states.quanta[state_ind]
]

plt.plot(time, pop_t[:, state_ind], label=state_qua)
plt.xlim([0, 20])

plt.title(
    f"OCS state population evolution in {round(peak_field/1e6, 1)} MV/cm THz field"
)
plt.xlabel("Time (ps)")
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend(fontsize=10, loc="center left", bbox_to_anchor=(1, 0.5))
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

Compute and plot alignment functions $\cos\beta$ and $\cos^2\beta$ ($\beta$ is the Euler angle between the molecular $z$ and the laboratory $Z$ axes)

```python
expval_cos_beta = np.einsum(
    "ti,it->t", np.conj(c_t), cos_beta.mat().dot(c_t.T), optimize="optimal"
)
expval_cos2_beta = (
    np.einsum("ti,it->t", np.conj(c_t), cos2_beta.mat().dot(c_t.T), optimize="optimal")
    + 1 / 3
)

plt.figure(figsize=(6, 4), dpi=300)
plt.plot(time, expval_cos_beta, linewidth=1.5, label="$\\cos\\theta$")
plt.plot(time, expval_cos2_beta, linewidth=1.5, label="$\\cos^2\\theta$")
plt.title(f"OCS alignment using {round(peak_field/1e6, 1)} MV/cm THz field")
plt.xlabel("Time (ps)")
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend(fontsize=10, loc="lower right")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
```

Compute and plot alignment functions $\cos\beta$, $\cos\beta_{2D}$, $\cos^2\beta$, and $\cos^2\beta_{2D}$, using Monte-Carlo approach.

```python
coefs = c_t[::10]

expval = []
nbatches = 10
for ibatch, c in enumerate(np.split(coefs, nbatches)):
    print(f"batch {ibatch+1}/{nbatches}")
    expval.append(states.mc_costheta(c.T))

mc_cos_beta, mc_cos2_beta, mc_cos_beta2d, mc_cos2_beta2d = np.concatenate(
    expval, axis=-1
)
```

```python
t = time[::10]

plt.figure(figsize=(6, 4), dpi=300)
plt.plot(t, mc_cos_beta, linewidth=1.5, label="$\\cos\\theta$")
plt.plot(t, mc_cos2_beta, linewidth=1.5, label="$\\cos^2\\theta$")
plt.plot(t, mc_cos_beta2d, linewidth=1.5, label="$\\cos\\theta_{2D}$")
plt.plot(t, mc_cos2_beta2d, linewidth=1.5, label="$\\cos^2\\theta_{2D}$")
plt.title(f"OCS alignment using {round(peak_field/1e6, 1)} MV/cm THz field")
plt.xlabel("Time (ps)")
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend(fontsize=10, loc="lower right")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
```
